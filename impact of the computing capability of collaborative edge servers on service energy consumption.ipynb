```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np

a=np.array([[12,12,12,14],[10,12,11,14],[8.5,12,10.5,14],[7,12,9,14],[6,12,8,14]])
df=DataFrame(a, columns=['IUCED','deep Q learning','Greedy search','Random search'],index= [3.19,6.38,9.57,12.76,15.95])

df.plot(kind='bar')

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.xlabel('The computing capability of edge servers(GHz)')
plt.ylabel('Average Energy consumption(J)')
 
plt.show()
```
