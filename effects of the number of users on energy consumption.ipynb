```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np

a=np.array([[6,7,10,10],[7,8,11,12],[8.5,9,12,13],[9,10,13,14],[10,12,14,16]])
df=DataFrame(a, columns=['IUCED','deep Q learning','Greedy search','Random search'],index= [10,60,110,160,210])

df.plot(kind='bar')

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.xlabel('Number of users')
plt.ylabel('Average Energy consumption(J)')
 
plt.show()

```
